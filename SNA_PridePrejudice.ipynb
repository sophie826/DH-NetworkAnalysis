```python
import csv
import re
from collections import defaultdict
```

```python
# Read the text file
with open('file.txt', 'r', encoding='cp1252') as f:
    text = f.read()
```

```python
# Define main characters mentioned in this chapter
characters = {
    'Elizabeth': ['Elizabeth', 'Miss Eliza', 'Miss Elizabeth', 'Miss Bennet'],
    'Mr. Darcy': ['Mr. Darcy', 'Darcy'],
    'Mr. Wickham': ['Mr. Wickham', 'Wickham', 'George Wickham'],
    'Mr. Bingley': ['Mr. Bingley', 'Bingley'],
    'Jane': ['Jane', 'her eldest sister', 'her sister'],
    'Charlotte Lucas': ['Charlotte Lucas', 'Charlotte', 'Miss Lucas'],
    'Mr. Collins': ['Mr. Collins', 'Collins'],
    'Miss Bingley': ['Miss Bingley'],
    'Mrs. Bennet': ['Mrs. Bennet', 'her mother'],
    'Mr. Bennet': ['Mr. Bennet', 'her father'],
    'Lydia': ['Lydia'],
    'Mary': ['Mary'],
    'Mr. Denny': ['Mr. Denny', 'Denny'],
    'Sir William Lucas': ['Sir William Lucas', 'Sir William'],
    'Lady Lucas': ['Lady Lucas'],
    'Lady Catherine': ['Lady Catherine', 'Lady Catherine de Bourgh'],
    'Miss De Bourgh': ['Miss De Bourgh'],
    'Mrs. Hurst': ['Mrs. Hurst']
}
```

```python
# Create reverse mapping for character detection
char_to_main = {}
for main_name, aliases in characters.items():
    for alias in aliases:
        char_to_main[alias] = main_name
```

```python
# Extract relationships based on co-occurance of names in one sentence
relationships = []

# Split the text into sentences using regular expressions
sentences = re.split(r'[.!?]+', text)

#Loop through all sentences and every sentence gets its own found_chars list
#If
for sentence in sentences:
    found_chars = []
    for alias, main_name in char_to_main.items():
        if alias in sentence:
            if main_name not in found_chars:
                found_chars.append(main_name)

    # If multiple characters in same sentence, they have a relationship
    if len(found_chars) >= 2:
        for i, char1 in enumerate(found_chars):
            for char2 in found_chars[i+1:]:
                relationships.append((char1, char2))

# Count relationship frequencies
relationship_counts = defaultdict(int)
for rel in relationships:
    relationship_counts[rel] += 1
print(relationship_counts)
```

```python
# Create nodes (all unique characters mentioned)
nodes = set()
for rel in relationship_counts.keys():
    nodes.add(rel[0])
    nodes.add(rel[1])
print(nodes)

# Write nodes to CSV
with open('nodes.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Id', 'Label'])
    for i, node in enumerate(sorted(nodes), 1):
        writer.writerow([i, node])
```

```python
# Create node to ID mapping
node_to_id = {node: i for i, node in enumerate(sorted(nodes), 1)}

# Write edges to CSV with weights (frequency of interaction)
with open('edges.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Source', 'Target', 'Weight', 'Type'])
    for (source, target), weight in sorted(relationship_counts.items(), key=lambda x: x[1], reverse=True):
        if weight >= 1:  # Only include relationships mentioned at least twice
            writer.writerow([node_to_id[source], node_to_id[target], weight, 'Undirected'])
```

```python
print(f"Extracted {len(nodes)} characters")
print(f"Extracted {len([w for w in relationship_counts.values() if w >= 1])} relationships (with weight >= 1)")
print("\nNodes saved to: nodes.csv")
print("Edges saved to: edges.csv")

# Display top relationships
print("\nTop 10 character interactions:")
for (char1, char2), count in sorted(relationship_counts.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"  {char1} <-> {char2}: {count} interactions")
```
